```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("movies.csv")
df.head()
```

```python
df1 = pd.read_csv("links.csv")
df1.head()
```

```python
df2 = pd.merge(df, df1, on="movieId", how="outer")
```

```python
df2.head()
```

```python
df2.shape
```

```python
df4 = pd.read_csv("ratings.csv")
df4.head()
```

```python
df4 = df4.drop(["userId", "timestamp"], axis = 1)
```

```python
df4[df4['movieId']==1] 
```

```python
df4 = df4.groupby(["movieId"]).mean()
```

```python
df4
```

```python
df5 = pd.merge(df2,df4, on= "movieId", how="outer")
```

```python
df5.set_index(['movieId'], inplace=True)
```

```python
df5["title"]= df5["title"].astype(str)
```

```python
df5 = df5.round({'rating':2})
```

```python
df5.to_excel("output.xlsx")
```

```python
df6 = df5['title'].str.split('(', 1, expand=True)
```

```python
df6 = df6.drop([0], axis =1)
```

```python
df6
```

```python
df6['Year'] = df6[1].str[-5:-1]
df6 = df6.drop([1], axis =1 )
```

```python
df7 = pd.merge(df5, df6, on="movieId", how="outer")
```

```python
df8 = df7['genres'].str.split('|', 4, expand=True)
```

```python
#df8 = df8.to_dict()
```

```python
df8 = df8.apply(pd.value_counts)
```

```python
df8['total'] = df8.sum(axis=1)
```

```python
indexNames = df8[ df8['total'] <= 40].index
 
# Delete these row indexes from dataFrame
df8.drop(indexNames , inplace=True)
```

```python
df8.drop([0,1,2,3,4], axis = 1, inplace = True)
```

```python
df8.to_dict()
```

```python
df7 = df7[['genres','rating']]
```

```python
df9 = df7['genres'].str.split('|', 4, expand=True)
```

```python
df10 = pd.merge(df9, df7, on="movieId", how="outer")
```

```python
df10 = df10.rename(columns = {0:"Genre", 1:"Genre1", 2:"Genre2", 3:"Genre3", 4:"Genre4"})
df10.drop(['genres'], axis = 1, inplace=True)
```

```python
df10 = df10[['Genre', 'rating']]
```

```python
df10.to_csv("review.csv")
```

```python
import csv, json

csvFilePath = "review.csv"
jsonFilePath = "review.json"

data = {}
with open (csvFilePath) as csvFile:
    csvReader = csv.DictReader(csvFile)
    for rows in csvReader:
        rating = rows["Genre"]
        data[rating] = rows
        
with open(jsonFilePath, 'w') as jsonFile:
    jsonFile.write(json.dumps(data))
```
